```python
##import the required libraries
import pandas as pd
import geopandas as gp
from shapely.geometry import Point
from shapely.geometry.polygon import LinearRing
from shapely.geometry import Polygon
import numpy as np
```

```python
##load ward data from open city website
##kml to converted csv
wards = pd.read_csv('././bengaluru_india.imposm-geojson/convertcsv.csv')
wards = gp.GeoDataFrame(wards)
df = gp.read_file('./bengaluru_india.imposm-geojson/bengaluru_india_buildings.geojson')
```

```python
##convert coordinates to linear ring
def getLinearRing (coordinates): 
    z=[]
    for i in np.array(coordinates.split(' ')):
        z.append([float(j) for j in i.split(',')])
    
    return LinearRing(z)
def iscontainedinWard(geo):
    for i,v in wards.iterrows():
        poly=Polygon(v['Polygon/outerBoundaryIs/LinearRing/coordinates'])
        if poly.contains(geo):
            
            return v['ExtendedData/Data/0/value']
        
```

```python
##manipulate data
for i, v in wards.iterrows(): 
    
    if v['Polygon/outerBoundaryIs/LinearRing/coordinates'] is np.nan:
       
        
        wards['Polygon/outerBoundaryIs/LinearRing/coordinates'][i]=wards['MultiGeometry/Polygon/0/outerBoundaryIs/LinearRing/coordinates'][i]+""

wards['Polygon/outerBoundaryIs/LinearRing/coordinates']=wards['Polygon/outerBoundaryIs/LinearRing/coordinates'].apply(getLinearRing)
```

```python
##example implementation

df['geometry']=df['geometry'].centroid
malldataframe=df[df['name'].str.contains('Mall')]

malldataframe['wards']=malldataframe['geometry'].apply(iscontainedinWard)
malldataframe[['name','wards']]
```

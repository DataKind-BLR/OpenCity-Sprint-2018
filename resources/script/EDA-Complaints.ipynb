```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib; matplotlib.style.use('ggplot')

from pandas import set_option

# inline plots
%matplotlib inline
```

```python
complaints = pd.read_csv('icmyc_complaints_june_2018.csv')
print("number of complaints raised: ", complaints.shape[0])
```

```python
ward_wise_complaints = complaints['ward_id'].value_counts(sort=True, ascending=False).to_frame()
```

```python
ward_wise_complaints.head(20).plot(kind='bar', title='Ward-wise complaints for 20 wards with most number of complaints')#, x='ward-id', y='complaint count')
```

```python
category_wise_complaints = complaints['category_id'].value_counts(sort=True, ascending=False).to_frame()
```

```python
category_wise_complaints.head(20).plot(kind='bar', title='Category-wise complaints 20 most common types of complaints')
```

```python
status_wise_complaints = complaints['complaint_status_title'].value_counts().to_frame()
```

```python
status_wise_complaints.head(20).plot(kind='bar', title='Status of complaints')
```

```python
civicAgency_wise_complaints = complaints['civic_agency_id'].value_counts().to_frame()
civicAgency_wise_complaints.head(20).plot(kind='bar', title='Civic Agencywise complaints')
```
